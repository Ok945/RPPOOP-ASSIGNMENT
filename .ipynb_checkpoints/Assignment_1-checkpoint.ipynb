```python
class Animal():
    def _init_(self):
        self.legs=2
        self.eyes=2  
        self.ears=2
    def geteyes(self):
        return self.eyes
    def types(self):  
        pass
    def habitat(self):
        pass
    def sound(self):
        pass
    def colour(self):
        pass
    def eat(self):
        pass
#single inheritance:
class Herbivores(Animal): 
    def types(self):
        print("single inheritance")
    def eat(self):         
        print("grass, herbs")
        
class carnivores(Animal):
    def child(self):
        print("Carnivores is child of parent Animal")
    def types(self):
        print("single inheritance")        
    def eat(self):
        print("Carnivores eat other animals")
        print("Animal->Carnivores")
        
class Omnivores(Animal):
    def eat(self):
        print("Plants and Animals")
    
        
class Dog(Omnivores):
    def _init_(self):    
         super(Dog,self)._init_()  
         self.legs=4
    def child(self):
        print("dog is child of parent omnivores")
    def types(self):
        print("inherits from Omnivores")
    def colour(self):
        print("White or brown")
    def eat(self):
        super().eat() 
        print("meat")
       
    def sound(self):
        print("bark")
    
class cat(Omnivores):
    def child(self):
        print("cat is child of parent omnivores")
    def _init_(self):
        super(cat,self)._init_()
        self.legs=4
    def types(self):
        print("inherits from Omnivores")
    def sound(self):
        print("Meow")
    def eat(self):
        print("Mouse") 
    def colour(self):
        print("White,brown,black")
class rabbit(Herbivores):
    def child(self):
        print("Rabbit is child of parent herbivores which is child of parent animal")
    def types(self):
        print("multilevel inheritance")
    def _init_(self):
        self.ears="Long"
    def getlegs(self):
        super(rabbit,self)._init_()
        print(self.legs)
    def sound(self):
        print("quirk")
        print("multiple levels are there :Animal->Herbivores->rabbit")
    def colour(self):
        print("White ,brown ,black")


        
        
#Multiple inheritance
class Mother():
    mname = ""
 
    def mother(self):
        print(self.mname)
class Father():
    fname = ""
 
    def father(self):
        print(self.fname)
class Child(Mother, Father):
    def child(self):
        print("It is child of mother and father ")
    def types(self):
        print("multiple inheritance")
    def parents(self):
        print("Father is", self.fname)
        print("Mother is", self.mname)
        print("Child is inherited from mother as well as father \n (Mother,Father)->child")

print("Choose type of inheritance \n 1)Single inherirtance \n 2)Multilevel inheritance \n 3)Multiple Inheritance \n 4)hierarchy inheritance \n 5)exit\n")
    
p = carnivores()
r = rabbit()

while(True):
    c = int(input("Enter choice: "))

    if(c==1):
        p.types()
        p.eat()
        print("\n")
    elif(c == 2):
        r.sound()
        r.colour()
        r.types()
        r.child()
        print("\n")
    elif(c==3):
        c = Child()
        c.types()
        c.fname="Rajesh"
        c.mname="Vaishali"
        c.parents()
        print("\n")
    elif(c == 4): 
        c = cat()
        c.types()
        d = Dog()
        d.types()
   
        print("Hierarchial inheritance both are inherits from omnivores\n Omnivores->(Cat, Dog)")
    else:
        break
```
