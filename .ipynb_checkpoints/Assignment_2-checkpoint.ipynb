```python
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        pass


class Dog(Animal):
    def sound(self):
        return "Woof"

class Cat(Animal):
    def sound(self):
        return "Meow"

class Cow(Animal):
    def sound(self):
        return "Moo"

class Shape:
    def __init__(self, sides):
        self.sides = sides

    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        super().__init__(0)
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Square(Shape):
    def __init__(self, side):
        super().__init__(4)
        self.side = side

    def area(self):
        return self.side * self.side


def animal_sound(animal):
    print(animal.sound())

def shape_area(shape):
    print(shape.area())

# main menu loop
while True:
    print("Choose an option:")
    print("1. CompileTime Polymorphism")
    print("2. RunTime Polymorphism")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        #  CompileTime Polymorphism example
        class temp:
            def add(self, a, b):
                return a + b

            def add(self, a, b, c):
                return a + b + c

        s = temp()
        try:
             print(s.add(5, 3))
        except:
            print("Method Overloading, functions having same name but different number of parameters")
            print("Here is the acceptable add function")
            print(s.add(2, 5, 8))
            


    elif choice == "2":
        # RunTime Polymorphism example
        print("Sound of Cow:")
        animal_sound(Cow("Moo"))
        print("object is of cow's class therefore function is called from cow's class")
        print("Sound of Dog:")
        animal_sound(Dog("Woof"))
        print("object is of dog's class therefore function is called from dog's class")
        print("Sound of Cat:")
        animal_sound(Cat("meow"))
        print("object is of cat's class therefore function is called from cat's class")      

    elif choice == "3":
        break

    else:
        print("Invalid choice. Please enter another option")

    
```
